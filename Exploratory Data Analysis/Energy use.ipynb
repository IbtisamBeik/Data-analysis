### Anaylzing the Energy use (kg of oil equivalent per capita)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#Read the Energy use csv file into a DataFrame:
df=pd.read_csv("C:\\Users\\ibsi\\Desktop\\Final project\\Data-analysis\\Datasets\\Energy consumption\\Energy use.csv")
df=df.rename(columns={"Country Name": "Country"})
df.head(5)
```

```python
df.dtypes
```

```python
# Checking for NaN values:
df.isna().sum()
# While there seems to be a lot of missing values, we are interested only in the last decade and therefore will check for missing values afterwards. 
```

```python
# We are interested in the data from the last 10 years only for our average Nominal GDP calculations:

df=df[["Country","2010", "2011", "2012", "2013", "2014", "2015"]]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
#Calculating the average Energy use from 2010 to 2015:

df["Avg_Energy_use"]=df[["2010", "2011", "2012", "2013", "2014", "2015"]].mean(axis=1)
```

```python
df1=df.nlargest(10,"Avg_Energy_use")
df1
```

```python
plt.subplots(figsize=(15, 10))
ax = df1.sort_values("Avg_Energy_use", ascending=False)
sns.set_color_codes("pastel")
sns.barplot(x="Avg_Energy_use", y="Country", data=ax,
                          palette="Reds_d")
sns.set(font_scale = 2)
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_2", transparent=True, bbox_inches="tight")
```
