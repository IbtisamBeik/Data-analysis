### Anaylzing the Nominal GDP (current US$)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#Read the Nomial GDPS csv file into a DataFrame:
df=pd.read_csv("C:\\Users\\ibsi\\Desktop\\Final project\\Data-analysis\\Datasets\\Economy and Growth\\GDP (current US$).csv")
df=df.rename(columns={"Country Name": "Country"})
df.head(5)
```

```python
# Checking for NaN values:
df.isna().sum()
# While there seems to be a lot of missing values, we are interested only in the last decade and therefore will check for missing values afterwards. 
```

```python
# We are interested in the data from the last 10 years only for our average Nominal GDP calculations:

df=df[["Country","2015", "2016", "2017", "2018", "2019"]]
```

```python
df.head()
```

```python
df.isna().sum()
```

For the year 2019 there are 37 missing values which is equivalent to 17% of the total data. However, we will keep the countries since they have valuable information for other years.  

```python
#Calculating the average GDPs (Nominal) from 2010 to 2019:

df["Avg_GDPs"]=df[["2015", "2016", "2017", "2018", "2019"]].mean(axis=1)
df.head()
```

```python
#sns.distplot(df["Avg_GDPs"]);
Top_50=df.nlargest(50,"Avg_GDPs")
sns.distplot(df["Avg_GDPs"], );
```

```python
df1=df.nlargest(5,"Avg_GDPs")
df1
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = df1.sort_values("Avg_GDPs", ascending=False)
sns.set_color_codes("pastel")
sns.barplot(x="Avg_GDPs", y="Country", data=ax,
                          palette="Blues_d").set(title = "GPD (Current US$) 2015-2019", xlabel = "", ylabel = "")
sns.set(font_scale = 2)
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_1", transparent=True, bbox_inches="tight")
```
