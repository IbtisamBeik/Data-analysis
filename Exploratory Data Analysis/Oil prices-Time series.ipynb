```python
from statsmodels.tsa.ar_model import AR
CrudeOil=df2["Crude oil average ($/bbl)"]

train, test = CrudeOil[:-59], CrudeOil[-59:]
model = AR(CrudeOil)
model_fit = model.fit(maxlag=1)
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
Estimated=pd.DataFrame({'observed':test, 'predicted':predictions})
Estimated
```

```python
import statsmodels.api as sm

X = Estimated["observed"]
Y = Estimated["predicted"]

Y=list(Y)

np.array(X), np.array(Y)
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```

```python
# Testing the Stationarity (H1) of the Oil prices timeseries:

from statsmodels.tsa.stattools import adfuller
p_value=adfuller(df2["Crude oil average ($/bbl)"])[1]
p_value
```

The p-value is above 0.05, which implies that our data is not stationary (The Null hypothesis).
