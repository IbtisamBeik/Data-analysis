```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# From the Fossil Fuel consumption DataFrame we need to extract the data for the top 3 consumers, those are: China, USA, India:

China=df1.loc[df1["Country"] == "China"]
USA=df1.loc[df1["Country"] == "United States of America"]
India=df1.loc[df1["Country"] == "India"]

frames = [China,USA,India]
result = pd.concat(frames)
del result["Country"]
df = pd.DataFrame(data=result)
df=df.reset_index(drop=True)
df = df.T
df = df.rename(columns={0: "China", 1: "USA", 2: "India"})
df
```

```python
from sklearn import linear_model
import statsmodels.api as sm
```

```python
# Multiple Linear Regression:

# with sklearn
X = df[["China","USA", "India"]]
Y = Prices["Crude oil average ($/bbl)"]

regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# with statsmodels

X = df[["China","USA", "India"]]
Y = Prices["Crude oil average ($/bbl)"]

Y=list(Y)

np.array(X), np.array(Y)
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```

```python
Estimated=pd.DataFrame({'observed':Y, 'predicted':predictions})
Estimated
```

```python
fig, ax = plt.subplots(nrows=1,
                       ncols=1,
                       figsize=(18,6))
a=sns.lineplot(data=Estimated, linewidth=3, ax=ax);
a.grid(False)
plt.xticks(color= "White", rotation=45);
plt.yticks(color= "White");
a.set_xlabel("Years", color= "White", size= 15);
a.set_ylabel("Crude oil prices ($/bbl)", color= "White", size= 15);
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_6", transparent=True, bbox_inches="tight");
```
